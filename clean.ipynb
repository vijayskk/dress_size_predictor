```python
import pandas as pd
```

```python
dataset = pd.read_csv('data.csv')
dataset
```

```python
dataset['size'].unique()
cols = {'XXS':0,'S':1,'M':2,'L':3,'XL':4,'XXL':5,'XXXL':6}

index = 0
for s in dataset['size']:
    dataset.loc[index,'size_in_num'] = int(cols[s])
    index += 1

dataset.dtypes
```

```python
dataset.astype({'size_in_num':int})
```

```python
dataset.dropna()
```

```python
dataset.to_csv('cleaned_data.csv')
```

```python
dataset.to_json('cleaned_data.json')
```
